```python
import folium
from folium.plugins import HeatMap
```

```python
from pyproj import Proj, transform


in_proj = Proj(init='epsg:3857')  # Projected coordinate system
out_proj = Proj(init='epsg:4326')  # Latitude/Longitude

# Filter for sewer backups
sewer_backups = df3[df3["Summary"] == "Sewer Back-ups (INTERNAL)"].copy()

# Convert coordinates
def convert_coords(x, y):
    lon, lat = transform(in_proj, out_proj, x, y)
    return lat, lon

sewer_backups[['Lat', 'Lon']] = sewer_backups.apply(lambda row: convert_coords(row['X'], row['Y']), axis=1, result_type='expand')

# Drop rows with NaN values in coordinates
sewer_backups = sewer_backups.dropna(subset=['Lat', 'Lon'])

# Extract location pairs for the heatmap
locations = sewer_backups[['Lat', 'Lon']].values.tolist()

# Create a Folium Map centered around Syracuse, NY
m = folium.Map(location=[43.0481, -76.1474], zoom_start=12)

# Add HeatMap layer
HeatMap(locations, radius=15, blur=10, min_opacity=0.5).add_to(m)

# Save and display the map
m.save("sewer_backup_heatmap.html")
```
